```python
s_class=[9,10,11,12]
boys=[60,50,90,85]
girls=[45,45,100,90]

bar1=[i+0.4 for i in s_class]
bar2=[i+0.2 for i in s_class]


plt.bar(s_class,boys,0.4,label="Boys")
plt.bar(bar1,girls,0.4,label="Girls")

plt.xticks(bar2,s_class)

plt.title("Student vs class")
plt.xlabel("Class")
plt.ylabel("No. of Students")
plt.legend()
```

# 3D GRAPH

```python
from mpl_toolkits import mplot3d
import numpy as np
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111, projection='3d')

x=[1,2,3,4,5,6,7,8,9]
y=[5,6,7,8,13,16,9,8,10]
z=np.zeros(9)

ax1.plot3D(x,y,z,color='brown',label="line")

ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
ax1.set_zlabel("z-axis")
ax1.legend()

plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111, projection='3d')

x=[1,2,3,4,5,6,7,8,9]
y=[5,6,7,8,13,16,9,8,10]
z=np.zeros(9)

dx=np.ones(9)
dy=np.ones(9)
dz=[1,2,4,6,8,5,8,3,9]

ax1.bar3d(x,y,z,dx,dy,dz,color='red')

ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
ax1.set_zlabel("z-axis")
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111, projection='3d')

x=[1,2,3,4,5,6,7,8,9]
y=[5,6,7,8,13,16,9,8,10]
z=[1,2,7,8,5,6,4,9,10]

w=[4,6,3,7,2,9,4,5,0]

ax1.scatter3D(x,y,z,cmap='Greens',c=w,s=100)
ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
ax1.set_zlabel("z-axis")

plt.show()
```
